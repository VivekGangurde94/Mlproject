```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
import matplotlib.pyplot as plt


# Assuming df is your DataFrame
plt.figure(figsize=(10, 6))

# Calculate cumulative sums for Team 1 and Team 2 points
cumulative_team1_points = df['team1 Points'].cumsum()
cumulative_team2_points = df['Team2 Points'].cumsum()

# Create a range of points representing the rows in the DataFrame
points = range(1, len(df) + 1)

# Plot cumulative points for each team
plt.plot(points, cumulative_team1_points,marker='o',label='Team 1 Cumulative Points', linestyle='-')
plt.plot(points, cumulative_team2_points,marker='o',label='Team 2 Cumulative Points', linestyle='-')

# Set x-axis ticks with step size 4
plt.xticks(range(1, len(df) + 1, 3))

plt.xlabel('Match')
plt.ylabel('Cumulative Points')
plt.title('Cumulative Points scored by Team 1 vs Team 2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping by 'RD' and summing the 'Bonus Point scored' column
grouped_data = df.groupby('RD')['Bonus Point scored'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(grouped_data, labels=grouped_data.index, autopct=lambda p: '{:.0f}'.format(p * sum(grouped_data) / 100), startangle=140, colors=['Orange', 'Blue'])
plt.title('Total Bonus Points Scored for UP and Gujarat')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grouping by 'RD' and summing the 'Bonus Point scored' column
grouped_data = df.groupby('RD')['Points By Raider'].sum()

# Plotting the graph
plt.figure(figsize=(8, 6))
grouped_data.plot(kind='bar', color=['blue', 'orange'])
plt.title('Total Bonus Points Scored for Raider (R) and Defense (D)')
plt.xlabel('Team (R: Raider, D: Defense)')
plt.ylabel('Total Bonus Points Scored')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

```
